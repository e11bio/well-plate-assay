```python
import zarr
from wellplate.elements import read_plate_xml
import numcodecs
meta_data, _, _ = read_plate_xml('C:/example_data/Johan-test.xml')
file_loc = 'C:/example_data/reprocessed.zarr'
z = zarr.open(file_loc)
#print(z.tree())
print(len(z['cells/intensity/well0/channel_gfp'][:]))
print(z['cells/masks/well0/channel_gfp'][:].max())
```

```python
z2 = zarr.open('C:/example_data/reprocessed2.zarr')
z2.require_group('cells')
z2.require_group('masks')
# meta data.
# masks.
channels = ['channel_dapi','channel_gfp','channel_rfp']
new_channels = ['365 nm','488 nm', '640 nm']
# masks
for i in range(len(channels)):
    for well in range(96):
        data = z[f'cells/masks/well{well}/{channels[i]}']
        z2.require_group(f'cells/masks/{well}/').create_dataset(new_channels[i], data = data, dtype='i2')
# intensity.
channels = ['channel_gfp','channel_rfp']
new_channels = ['488 nm', '640 nm']
for i in range(len(channels)):
    for well in range(96):
        data = z[f'cells/intensity/well{well}/{channels[i]}']
        z2.require_group(f'cells/intensity/{well}/').create_dataset(new_channels[i], data = data, dtype='f')
```

```python
print(z2.tree())
```
